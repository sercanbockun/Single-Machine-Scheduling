```python
###################
"""
PLEASE REALISE THERE ARE TWO WAYS OF WRITING THE CONSTRAINTS WHICH IS NEEDED TO BLOCK OVERLAPPINGS. I WROTE BOTH OF THEM AND 
COMMENTED OUT THE ONE WHICH WAS GIVEN IN THE CLASS. 

COMPUTATIONAL TIME (APPROXIMATELY) USING THE:

   - ALTERNATIVE CONSTRAINT : 30.9 MS 
   
   - CONSTRAINT GIVEN IN THE CLASS : 42.6 MS

"""
###################



from pulp import *

process_time = {'1':2, '2':2, '3':1, '4':2, '5':3}
due_time = {'1':3, '2':3, '3':1, '4':9, '5':4}
jobs = [1,2,3,4,5]

M = 100  

model = LpProblem(name = "single machine scheduling", sense = LpMinimize)

x = {(j,t): LpVariable(name=f"x{j}{t}", cat="Binary") for j in range(1,6) for t in range(1,11)}
y = {j: LpVariable(name=f"y{j}", cat="Binary") for j in range(1, 6)} # whether job missed due date or not. If job missed y(j)=1

for j in range(1,6):
    model += lpSum(x[j,t] for t in range(1,12-process_time[f"{j}"])) ==1 # every job must start at a time which is certainly before 11-processtime

    ##############################################################
#for job in range(0,5):
 #   for time in range(1,11):
  #      for job2 in (jobs[:job]+jobs[job+1:]):
   #         for t in range(time, min(11,time+ process_time[f"{job+1}"])):
    #            model+= x[job+1,time]+x[job2, t]<=1         # THIS IS THE CONSTRAINT WHICH WAS GIVEN IN THE CLASS
############################################################## PLEASE REALISE IF YOU UNCOMMENT THIS SECTION AND COMMENT OUT THE BELOW FOR LOOP SAME RESULT WÄ°LL BE DERIVED. 
#UPPER COMMENTED SECTION IS THE CONSTRAINT WHICH WAS GIVEN IN THE CLASS.



        
for time in range(1,11):
    model += lpSum(x[j,t] for j in range(1,6) for t in range(max(time +1- process_time[f"{j}"],1),time+1 )) <=1 
   #This is the constraint which make sures there's at most one
    #process at a given 'time'. THIS CONSTRAINT WAS NOT GIVEN IN THE CLASS AND DIFFERENT FROM IT.


for j in range(1,6):
    for t in range(1,11):
        model += M*(y[j])>= x[j,t]*(process_time[f"{j}"] + t - due_time[f"{j}"]) -1

model += lpSum(y.values())

%time status = model.solve()
print(f"status: {model.status}, {LpStatus[model.status]}")
print(f"objective: {model.objective.value()}")

for var in model.variables():
    print(f"{var.name}: {var.value()}")
```
